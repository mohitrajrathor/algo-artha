# Predicting daily price

```python
import yfinance as yf
import pandas as pd 
import numpy as np 
```

```python
nifty50 = pd.read_csv("../data/nifty-500.csv")
nifty50_symbols = (nifty50['Symbol'] + ".NS").to_list()
```

```python
# downloading nifty 50 data from yfinance

nifty50_df = yf.download(tickers=nifty50_symbols, start="2014-01-01", period='1d')
nifty50_df
```

```python
nifty50_df = nifty50_df.stack()
nifty50_df
```

```python
df = nifty50_df.copy()
df.loc['2024-06-12']
```

## Technical analysis indicator addition

```python
import pandas_ta as ta 
```

```python
# calculate moving averages
nifty50_df['EMA-5'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: ta.ema(x, period=5))
nifty50_df['EMA-20'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: ta.ema(x, period=20))
nifty50_df['EMA-50'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: ta.ema(x, period=50))
nifty50_df['EMA-200'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: ta.ema(x, period=200))

# calculate RSI
nifty50_df['RSI'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: ta.rsi(x, length=14))

# calculate next open
nifty50_df['Next Open'] = nifty50_df.groupby(level='Ticker')['Open'].transform(lambda x: x.shift(-1))

# calculate target
nifty50_df['target_1d'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: x.shift(-1))
nifty50_df['target_10d'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: x.shift(-10))
nifty50_df['target_30d'] = nifty50_df.groupby(level='Ticker')['Adj Close'].transform(lambda x: x.shift(-30))

# calculate change
nifty50_df['change_1d'] = ((nifty50_df.groupby(level='Ticker')['target_1d'].transform(lambda x: x) - nifty50_df.groupby(level='Ticker')['Open'].transform(lambda x: x)) / nifty50_df.groupby(level='Ticker')['Open'].transform(lambda x: x)) * 100
nifty50_df['change_10d'] = ((nifty50_df.groupby(level='Ticker')['target_10d'].transform(lambda x: x) - nifty50_df.groupby(level='Ticker')['Open'].transform(lambda x: x)) / nifty50_df.groupby(level='Ticker')['Open'].transform(lambda x: x)) * 100
nifty50_df['change_30d'] = ((nifty50_df.groupby(level='Ticker')['target_30d'].transform(lambda x: x) - nifty50_df.groupby(level='Ticker')['Open'].transform(lambda x: x)) / nifty50_df.groupby(level='Ticker')['Open'].transform(lambda x: x)) * 100
```

### Training regressiom model to predict next days close price 



```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge


features = ['EMA-5', 'EMA-20', 'EMA-50', 'EMA-200', 'RSI', 'Volume', 'Next Open', 'Adj Close']

X = nifty50_df[features]
y_1d = nifty50_df['target_1d']
y_10d = nifty50_df['target_10d']
y_30d = nifty50_df['target_30d'] 


# model training to predict one day return.
X_train, X_test, y_train, y_test = train_test_split(X, y_1d, test_size=0.33, random_state=42)

model_1d = LinearRegression()
lasso = Lasso(alpha=0.1)
model_1d.fit(X_train, y_train)
lasso.fit(X_train, y_train)

y_pred = lasso.predict(X_test)

print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
# model to predict 10 days return 
X_train, X_test, y_train, y_test = train_test_split(X, y_10d, test_size=0.33, random_state=42)

model_10d = LinearRegression()
model_10d.fit(X_train, y_train)

lasso2 = Lasso(alpha=1)
lasso2.fit(X_train, y_train)

y_pred = model_10d.predict(X_test)

print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
# model to predict 30 days return 
X_train, X_test, y_train, y_test = train_test_split(X, y_30d, test_size=0.33, random_state=42)

model_30d = LinearRegression()
model_30d.fit(X_train, y_train)

lasso3 = Lasso(alpha=0.1)
lasso3.fit(X_train, y_train)


y_pred = model_30d.predict(X_test)

print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```
