### Login method
create ./config/cred.yaml file in the root folder, the cred.yaml file should contains followings:
```yaml
user    : <user-id>
pwd     : <password>
totptoken : <totp - token>
vc      : <vendor - code>
apikey  : <api - key>
imei    : <imei>
```

*For more information about shoonya api *[Github Page](https://github.com/Shoonya-Dev/ShoonyaApi-py)*.

````python
# create a api instance
api = ShoonyaApiPy()

# raed cred.yaml data using yaml file.
with open('config/cred.yml') as f:
    cred = yaml.load(f, Loader=yaml.FullLoader)

# now loaging to api as follows.
ret = api.login(
    userid=cred['user'],
    password=cred['pwd'],
    twoFA=pyotp.TOTP(cred['totptoken']).now(),
    vendor_code=cred['vc'],
    api_secret=cred['apikey'],
    imei=cred['imei']
)

ret
````

````python
def cross_time(tg_time:str):
    ''' 
    param -> tg_time 
                pass time in '<year>-<month>-<day> <hour>:<minute>:<second>'
    '''
    curr_time = dt.datetime.now()
    target_time = dt.datetime.now().strptime(tg_time, "%Y-%m-%d %H:%M:%S")

    if curr_time > target_time: return True
    else: return False
````

````python
def check_brakout(sym:str)->bool:
    curr_time = dt.datetime.now()
    start = curr_time - dt.timedelta(days=5)
    ret = api.get_time_price_series(
        exchange="NSE", token=str(nse_master['Token'].loc[sym]), 
        starttime=start.timestamp(), 
        interval="5" 
    )
    data = pd.DataFrame(ret)
    print(data)
    _5_min_high = float(data[data['time'] == "01-01-2024 09:15:00"]['intc'])
    curr_close = float(data['intc'].loc[0])
    prev_close = float(data['intc'].loc[1])
    if prev_close < _5_min_high < curr_close:
        return True
    else : return False
    
````
